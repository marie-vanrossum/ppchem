```python
import matplotlib.pyplot as plt

# Define your data
y_values = [3.16, 3.01, 3.06]
x1_values = [500, 460, 307]
x2_values = [100, 85, 75]

# Create figure and axes
fig, ax1 = plt.subplots()

# Plot the first set of data
ax1.plot(x1_values, y_values, 'o', label='In function of irradiancy', color='blue')
ax1.set_xlabel('Irradiancy [W/m^2]')
ax1.set_ylabel('Efficiency [%]', color='cornflowerblue')

# Create a twin axes sharing the same y-axis
ax2 = ax1.twiny()

# Plot the second set of data
ax2.plot(x2_values, y_values, 'o', label='In function of light intensity', color='red')
ax2.set_xlabel('Light intensity [%]')
ax2.tick_params(axis='x', colors='palevioletred')

# Add legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='upper left')

plt.title('Efficiency in function of light intensity and irradiancy')
plt.grid(True)
plt.ylim(2, 3.5)
plt.show()
```
