```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Generate some sample data
x_data = np.array([6, 7, 8, 9, 10])
y_data = np.array([2.287333333, 2.413, 2.645, 3.074, 3.859])

# Define the exponential function
def exponential_func(x, k, a, b):
    return k * np.exp(a * x) + b

# Perform exponential regression
params, covariance = curve_fit(exponential_func, x_data, y_data)

# Plot the original data
plt.scatter(x_data, y_data, color='cornflowerblue', label='Data')

# Plot the regression curve
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = exponential_func(x_fit, *params)
plt.plot(x_fit, y_fit, color='violet', label='Exponential Regression Curve')


plt.xlabel('number of carbons')
plt.ylabel('Retention time [min]')
plt.title('Exponential Regression for Determination of Death Time')
plt.legend()
plt.grid(True)
plt.show()

# Calculate residuals
residuals = y_data - exponential_func(x_data, *params)

# Calculate total sum of squares
total_sum_squares = np.sum((y_data - np.mean(y_data)) ** 2)

# Calculate sum of squares of residuals
residual_sum_squares = np.sum(residuals ** 2)

# Calculate R-squared value
r_squared = 1 - (residual_sum_squares / total_sum_squares)

print("R-squared value:", r_squared)
```

```python
import numpy as np
from scipy.optimize import curve_fit

# Generate some sample data
x_data = np.array([6, 7, 8, 9, 10])
y_data = np.array([2.287333333, 2.413, 2.645, 3.074, 3.859])

# Define the exponential function with a constant term
def exponential_func(x, a, b, k):
    return a * np.exp(b * x) + k

# Perform exponential regression
params, covariance = curve_fit(exponential_func, x_data, y_data)

params_std = np.sqrt(np.diag(covariance))

# Equation of the regression
equation = f'y = {params[0]:.6f} ± {params_std[0]:.6f} * e^({params[1]:.6f} ± {params_std[1]:.6f} * x) + {params[2]:.6f} ± {params_std[2]:.6f}'
print("Regression Equation:")
print(equation)
```

```python
import numpy as np
from scipy import stats

# Data
x = np.array([0.551276, 0.646394, 0.960922, 1.440661, 1.936538])
y = np.array([0.87295894, 1.025170146, 1.505157573, 2.272100207, 3.112635658])

# Linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Equation of the linear regression line
equation = f'y = {slope:.2f}x + {intercept:.2f}'

# Uncertainties
slope_uncertainty = std_err
intercept_uncertainty = std_err

# Equation with uncertainties
equation_with_uncertainties = f'y = {slope:.3f} ± {slope_uncertainty:.3f}x + {intercept:.3f} ± {intercept_uncertainty:.3f}'

r_squared = r_value**2

print("R^2 value:", r_squared)

print("Equation of the linear regression line:", equation)
print("Equation with uncertainties:", equation_with_uncertainties)
```
